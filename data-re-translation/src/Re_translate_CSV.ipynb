```
import numpy as np
import pandas as pd
import googletrans
from googletrans import Translator
```

```
df = pd.read_csv('https://raw.githubusercontent.com/shunjid/cn-covid-storybook/master/data-wrangling/out/covid-en-0.5.csv')
```

```
df.head()
```

```
df_description = df['Event'].tolist()
```

```
for item in df_description[:]: # make a copy of x
    if 'On' in item or 'January' in item or 'February' in item:
        df_description.remove(item)
```

```
len(df_description)
```

```
for item in df_description[:]: # make a copy of x
    if 'December' in item:
        df_description.remove(item)
```

```
len(df_description)
```

```
df_description[2000:4000]
```

```
desc_set = set(df_description)
```

```
len(desc_set)
```

```
desc_set
```

```
translator = Translator()
translations = {}
i = 0

for element in desc_set:
    # add translation to the dictionary
    translations[element] = translator.translate(element).text
    i = i + 1
    print('t => {}'.format(i))


print('Done')
```

```
len(translations)
```

```
translations
```

```
# modify all the terms of the data frame by using the previously created dictionary
df.replace(translations, inplace = True)

# check translation
df.head()
```

```
df['Event'].replace(np.nan, '', inplace=True)
df_event = df['Event'].tolist()
```

```
event_set = set(df_event)
len(event_set)
```

```
event_set
```

```
translator = Translator()
translations = {}
i = 0

for element in event_set:
    # add translation to the dictionary
    translations[element] = translator.translate(element).text
    i = i + 1
    print('t => {}'.format(i))


print('Done')
```

```
len(translations)
```

```
# modify all the terms of the data frame by using the previously created dictionary
df.replace(translations, inplace = True)

# check translation
df.head()
```
